```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/DSML datasets/iris.csv")
```

```python
df.head()
```

**Returns the first 5 lines of the file**

```python
  plt.hist(df['sepal.length'],edgecolor='red')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

**Describes the count of sepal length distributed to each of the sepal length class**

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='sepal.width',y='variety',data=df)
plt.xlabel('sepal width')
plt.ylabel('variety')
plt.title('Quartile Plot')
plt.show()
```

**Plot shows the range of values present for each 3 features based on the sepal width**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['sepal.length'])
plt.xlabel('sepal length')
plt.ylabel('Density')
plt.title('Distribution Chart')
plt.show()
```

**Shows the range of sepal lengths with respective to its density**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal.length',y='sepal.width',hue='variety',style='variety',data=df,s=100)
plt.xlabel('sepal length')
plt.ylabel('sepal.width')
plt.title('Scatter Plot- sepal length vs sepal width by species')
plt.legend(title='species',loc='upper left')
plt.show()
```

**Show the data points with respect to sepal length and width of 3 species**

```python
sns.pairplot(df,hue='variety',markers=['o','s','d'],diag_kind='kde',height=2)
plt.suptitle('Scatter Matrix of iris dataset by species',y=1.02)
plt.show()
```

**Indicates each of the features in the dataset with each other features specified by species**

```python
  x=df['sepal.length']
  y=df['sepal.width']
  size=df['petal.length']*10
  plt.figure(figsize=(8,5))
  plt.scatter(x,y,s=size,alpha=0.7,c='red',edgecolors='black')
  plt.xlabel('sepal length')
  plt.ylabel('sepal width')
  plt.title('Bubble Chart- sepal length vs sepal width(petal length as bubble size)')
  plt.show()
```

** Indicates the sepal length vs sepal width  of the data attributes**

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['sepal.length'],shade=True,color='red')
plt.xlabel('sepal length')
plt.ylabel('Density')
plt.title('Density chart of sepal length')
plt.show()

```

**Shows the density of the faeture sepal length **

```python
mean_sepal_length=df['sepal.length'].mean()
df['deviation']=df['sepal.length']-mean_sepal_length
plt.figure(figsize=(8,5))
plt.bar(df['variety'],df['deviation'],color='green',edgecolor='red')
plt.xlabel('Species')
plt.ylabel('Deviations')
plt.title('Deviation Chart')
plt.show()
```

**deviations of each species based on sepal length**

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(8,5))
andrews_curves(df,'variety',colormap='viridis')
plt.legend(title='variety',loc='upper left')
plt.title('Andrews Curves')
plt.show()
```
