```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('/content/drive/MyDrive/DSML datasets/Salary_Data - Salary_Data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
X=data.drop('Salary',axis=1)
Y=data.drop('YearsExperience',axis=1)
```

```python
X.head()
```

```python
Y.head()
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=.2,random_state=2)
```

```python
# x_train=np.reshape(x_train,(-1,1))
# x_test=np.reashpe(x_test,(-1,1))
# y_train=np.reshape(y_train,(-1,1))
# y_test=np.reashpe(y_test,(-1,1))
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_train)
test_pred=model.predict(x_test)
```

```python
mse_train=mean_squared_error(y_train,prediction)
mse_test=mean_squared_error(test_pred, y_test)
print("Train mse",mse_train)
print("Test mse",mse_test)
```

```python
train_acc=r2_score(y_train,prediction)
test_acc=r2_score(test_pred,y_test)
print("Training Accuracy",train_acc)
print("Testing Accuracy",test_acc)
```

```python
mae_train=mean_absolute_error(y_train,prediction)
mae_test=mean_absolute_error(test_pred, y_test)
print("Train mae",mae_train)
print("Test mae",mae_test)
```
