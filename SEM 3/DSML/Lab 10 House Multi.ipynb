```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("/content/drive/MyDrive/DSML datasets/Housing - Housing.csv")
```

```python
lb=LabelEncoder()
```

```python
data['mainroad']=lb.fit_transform(data['mainroad'])
data['guestroom']=lb.fit_transform(data['guestroom'])
data['basement']=lb.fit_transform(data['basement'])
data['hotwaterheating']=lb.fit_transform(data['hotwaterheating'])
data['airconditioning']=lb.fit_transform(data['airconditioning'])
data['prefarea']=lb.fit_transform(data['prefarea'])
data['furnishingstatus']=lb.fit_transform(data['furnishingstatus'])
```

```python
data.head()
```

```python
X=data.drop(['price'],axis=1)
Y=data['price']
```

```python
x_train,x_test, y_train, y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
train_data_prediction=model.predict(x_train)
accuracy=metrics.r2_score(train_data_prediction,y_train)
mae=metrics.mean_absolute_error(train_data_prediction,y_train)
mse=metrics.mean_squared_error(train_data_prediction,y_train)
print("Accuracy= ",accuracy)
print("Mean Absolute Error= ",mae)
print("Mean Squared Error= ",mse)
```

```python
test_data_prediction=model.predict(x_test)
test_accuracy=metrics.r2_score(test_data_prediction,y_test)
test_mae=metrics.mean_absolute_error(test_data_prediction,y_test)
test_mse=metrics.mean_squared_error(test_data_prediction,y_test)
print("Accuracy= ",test_accuracy)
print("Mean Absolute Error= ",test_mae)
print("Mean Squared Error= ",test_mse)
```
