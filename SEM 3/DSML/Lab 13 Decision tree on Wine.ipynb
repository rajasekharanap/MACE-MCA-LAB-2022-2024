```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/DSML datasets/WineQuality - WineQuality.csv')
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['type'].unique()
encoder=LabelEncoder()
data['type']=encoder.fit_transform(data['type'])
```

```python
featurenames=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','pH','sulphates',]
for fn in featurenames:
  data[fn].fillna(data[fn].mean(), inplace=True)

data.isnull().sum()
```

```python
data['quality'].unique()
```

```python
X=data.drop('quality',axis=1)
X.head(3)
```

```python
Y=data['quality']
Y.head(3)
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=.2,random_state=2)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
train_acc=metrics.accuracy_score(train_pred,y_train)
print("Accuracy= ",train_acc)
```

```python
test_pred=model.predict(x_test)
test_acc=metrics.accuracy_score(test_pred,y_test)
print("Accuracy= ",test_acc)
```
