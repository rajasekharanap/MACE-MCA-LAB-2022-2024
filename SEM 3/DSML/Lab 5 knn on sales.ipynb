```python
#Import libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
```

```python
train_data=pd.read_csv('/content/drive/MyDrive/DSML datasets/Sales_Train - Sales_Train.csv')
test_data=pd.read_csv('/content/drive/MyDrive/DSML datasets/Sales_Test - Sales_Test.csv')
```

```python
test_data.fillna(0, inplace=True)
```

```python
train_data['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'}, inplace=True)
train_data['Item_Fat_Content'].value_counts()
```

```python
test_data['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'}, inplace=True)
test_data['Item_Fat_Content'].value_counts()
```

```python
train_data['Item_Type'].value_counts()
```

```python
#encode categorical data
train_data=pd.get_dummies(train_data,columns=['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
test_data=pd.get_dummies(test_data,columns=['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
```

```python
#split the train data to features and target
X=train_data.drop(['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'],axis=1)
Y=train_data['Item_Outlet_Sales']
```

```python
#create the KNN model
model=KNeighborsRegressor(n_neighbors=3)
```

```python
model.fit(X, Y)
```

```python
# Use the model to make predictions on the test data
X_test = test_data.drop(['Item_Identifier', 'Outlet_Identifier'], axis=1)
test_predictions =model.predict(X_test)
```

```python
training_data_prediction = model.predict(X)
score_train = str(model.score(X, Y))
r2_train = metrics.r2_score(Y, training_data_prediction)
r2_train_mse= metrics.mean_squared_error(Y, training_data_prediction)
r2_train_mae= metrics.mean_absolute_error(Y, training_data_prediction)
print(score_train)
print('R_2 Score value= ', r2_train)
print('MAE value= ', r2_train_mae)
print('RMSE value= ',np.sqrt(r2_train_mse))
```
