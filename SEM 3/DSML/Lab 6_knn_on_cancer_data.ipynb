```python
# separating features and labels
X = cancer.drop(columns=['diagnosis','id'], axis=1)
Y = cancer['diagnosis']
```

```python

Y.head()
```

```python
X.head()
```

```python
X_train, X_test,Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(cancer.shape, X.shape, Y.shape, X_train.shape, X_test.shape,Y_train.shape, Y_test.shape)
```

```python
# model training
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train, Y_train)
```

```python
sample = [[15.78,17.89,103.6,781,0.0971,0.1292,0.09954,0.06606,0.1842,0.06082,0.5058,0.9849,3.564,54.16,0.005771,0.04061,0.02791,0.01282,0.02008,0.004144,20.42,27.28,136.5,1299,0.1396,0.5609,0.3965,0.181,0.3792,0.1048]]
predict = model.predict(sample)
if predict[0]=='B':
  print('Benign')
if predict[0]=='M':
  print('Malignant')
```
