```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('/content/drive/MyDrive/DSML datasets/iris.csv')
```

```python
data.describe()
```

```python
data.info()
```

```python
data.shape
```

```python
data.head()
```

```python
encoder=LabelEncoder()
data['variety']=encoder.fit_transform(data['variety'])
```

```python
X=data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
X.head()
```

```python
Y=data['variety']
Y.head()
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=.2,random_state=42)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(X_train,Y_train)
```

```python
pred_train=model.predict(X_train)
accuracy=metrics.accuracy_score(Y_train,pred_train)
print("Train Accuracy- ",accuracy)
```

```python
pred_test=model.predict(X_test)
accuracy=metrics.accuracy_score(Y_test,pred_test)
print("Test Accuracy- ",accuracy)
```
